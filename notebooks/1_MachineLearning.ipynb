# Визуализация

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

movie_data = pd.read_csv('../data/CSM_snake_case.csv')
movie_numeric_df = pd.concat([movie_data['Views'],
                       movie_data['Ratings'],movie_data['Genre'],
                       movie_data['Likes'],movie_data['Budget'], movie_data['Aggregate Followers'],
                        movie_data['Screens']], axis=1,
                       keys=['Views', 'Ratings','Genre','Likes','Budget','Aggregate Followers','Screens'])
movie_numeric_df.hist(bins=20, figsize=(20,20))
plt.show()
```

Сделанная выше визуализация показывает распределение просмотров, рейтингов, жанров, лайков, бюджета, кол-во кинотеатров из общего количества кинофильмов в выборке
